# Set x-axis, y-axis, and title

```python
import matplotlib.pyplot as plt
days = ["Madgaon","Kolhapur","Panjim","Sawantwadi","Kankavali","Rajapur","Hathkhamba","Sangli","Velleapur","Kavathemahakal"]
temperature = [36.6, 37,37.7,39,40.1,43,43.4,45,45.6,40.1,44,45,46.8,47,47.8]
plt.plot(days, temperature)
plt.title("PMT Status")
plt.xlabel("NMT")
plt.ylabel("PMT Pending")
plt.show()
```
